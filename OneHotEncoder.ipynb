```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()
```

```python
ohe.fit_transform(df[['Colour','Country']]).toarray()
```

```python
fea_array=ohe.fit_transform(df[['Colour','Country']]).toarray()
```

```python
ohe.categories_
```

```python
feature_labels=ohe.categories_
```

```python
feature_labels=np.array(feature_labels).ravel()
```

```python
feature_labels
```

```python
df1=pd.DataFrame(fea_array,columns=feature_labels)
```

```python
df1.head(10)
```

```python
df2=pd.concat([df,df1],axis=1)
```
